### These are just for random queries and introduction

```python
q_words = ['what',"what's", 'which', 'who', 'where', 'why', 'when', 'how']
greet = ['hello', 'hi', 'hey', 'hi there', 'whats up']
endings = ['bye', 'good bye', 'thank you', 'thanks', 'peace out']
itself = ['you', 'your','yourself', 'yours', "you're"]
thank = ['thank you sir', 'thanks boss', 'welcome', "you're welcome", 'have a nice day', 'nice to meet you']
Tiger_words = ["tiger", "tigers","tiger's", "lions", "they","cat","cats", "their", "them", "cub", "cubs"]
responces = {'me':["i am an AI chatbot made by Mr. shubhendu ghosh", "I am a virtual asistant that can help you with tigers", " I am a machine Intelligence made for helping humans know about tigers", "i can help you with tigers", "I can help you with your tiger quiz", "You can ask me about Tigers and i will answer you", "I am a Tiger expert"],
             'open':['Hi sir, How can i help you?', "welcome boss, what's up", "hello there. may i help you ?","Do you need someone to talk to?", "what can i do for you, sir", "welcome sir, would you like to know about tigers? i am an expert" ],
             'end' : thank,
             'sorry': ["I'm sorry sir, i could'nt get what you just said", "I'm sorry sir, would you please ask again, this time more specific.", "I failed you sir", "My bad, I don't know that sir", "I am not programmed for this sir, please ask me about tiger related things", "I don't have an answer for this sir, sorry!" ],
             'name':["I don't have a name sir", "I am not human, you can call me chatbot", "whatever you'll call me", "call me as you want", "Nobody named me yet", "i don't have a specific name yet"]
             }
#len(responces['end'])
```

### These functions are quite handy

```python
def remove_ques(text):
  tok = text.split()
  for i in q_words:
    if i in tok:
      idx = tok.index(i)
      tok.pop(idx)
      break
  text = ' '.join(tok)
  return text


#remove_ques("what's is your name")
def have_you(text):
  res = False
  tok = text.split()
  for i in itself:
    if i in tok:
      res = True
      break
  return res

def greeting(text):
  res = False
  tok = text.split()
  for i in greet:
    if i in tok:
      res = True
      break
  return res


def have_tiger(text):
  res = False
  tok = text.split()
  for i in Tiger_words:
    if i in tok:
      res = True
      break
  return res

def ending(text):
  res = False
  tok = text.split()
  for i in endings:
    if i in tok:
      res = True
      break
  return res
```

```python
import random
```

### This function will return our answers

```python
def response(text):
  text = text.lower()
  #text = remove_ques(text)
  if greeting(text) == True:
    p = random.randint(0, len(responces['open']))
    print(responces['open'][p])
  elif have_you(text) == True:
    p = random.randint(0, len(responces['me']))
    print(responces['me'][p])
  elif ending(text) == True:
    p = random.randint(0, len(responces['end']))
    print(responces['end'][p])
  elif have_tiger(text) == True:
    answer(text)
  else:
    p = random.randint(0, len(responces['sorry']))
    print(responces['sorry'][p])
```

### Let's test it

```python
query = " "
while query != '0':
  query = input("You : ")
  print("Chatbot : ", end = "")
  response(query)
```
